```python
import pandas as pd
from sklearn.cluster import KMeans
import plotly.express as px

# Step 1: Load the CSV data into a pandas DataFrame
data = pd.read_csv("csvdata/datafullnight2_SE.csv")
```

```python
data = pd.read_csv("csvdata/datafullnight2_SE_waso0.25.csv")

# Step 2: Create an interactive 3D scatter plot using Plotly
fig = px.scatter_3d(data, x="WASO_min0.25original", y="AUTO_SE", z="N2 (min)",
                    color="LTDP10", size_max=10)

fig.update_layout(scene=dict(xaxis_title="WASO_min0.25original", yaxis_title="AUTO_SE",
                             zaxis_title="N2 (min)"),
                  title="3D Scatter Plot",
                  showlegend=True)

# Step 3: Increase the height of the plot (optional)
fig.update_layout(height=1000)  # Set the desired height (in pixels)

# Step 4: Show the plot
fig.show()
```

```python


# Step 2: Select columns for clustering (all but the target columns.)
columns_for_clustering = data.drop(["nsrrid", "LTDP10", "REST10"], axis=1)


# Step 3: Run K-means clustering with n=5 clusters
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
data["cluster_label"] = kmeans.fit_predict(columns_for_clustering)


# Step 4: Calculate average stats per cluster
cluster_stats = data.groupby("cluster_label").mean()


pd.set_option('display.max_columns', None)
pd.set_option('expand_frame_repr', False)

# Step 5: Print average stats per cluster
print("Average Stats Per Cluster:")
print(cluster_stats)

# Step 6: Calculate most dominant statistic per column
dominant_stats = data.groupby("cluster_label").apply(lambda x: x.mode().iloc[0])

# Step 7: Print most dominant statistic per column
print("\nMost Dominant Statistic Per Column:")
print(dominant_stats)

# Step 4: Create an interactive 3D scatter plot using Plotly
fig = px.scatter_3d(data, x='column1', y='column2', z='column3',
                    color="cluster_label", size_max=10, symbol="cluster_label")

fig.update_layout(scene=dict(xaxis_title="Column 1", yaxis_title="Column 2",
                             zaxis_title="Column 3"),
                  title="K-means Clustering (n=5) - Patient Data",
                  legend_title="Cluster Label",
                  showlegend=False)

# Step 5: Increase the height of the plot
fig.update_layout(height=600)  # Set the desired height (in pixels)

# Step 6: Show the plot
fig.show()
```
